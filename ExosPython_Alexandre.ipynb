Exemple : fonction(42.75, 52.23) doit renvoyer 43 45 47 49 51

```python
a = 5
b = 15
liste_impairs = []
liste_pairs = []

def creer_liste_impairs_et_pairs(a,b):

    for i in range(a,b+1):
        if (i%2 != 0):
            liste_impairs.append(i)
        else :
            liste_pairs.append(i)
    
creer_liste_impairs_et_pairs(a,b)

liste_impairs
# liste_pairs  <= petit bonus: on peut aussi afficher la liste des nombres pair, avec liste_pairs
```

```python
import random
n = random.randint(1,100)
essais = 5
p =  int(input("Entrez un nombre de 1 à 100 (5 essais restants): "))

while essais > 1 :

    if p > n :
        print(f"{p} est TROP GRAND! Essaye encore ({essais-1} essais restants):")
        p =  int(input())
    elif p < n :
        print(f"{p} est trop petit ! Essaye encore({essais-1} essais restants):")
        p =  int(input())
    elif p == n: 
        print(f"BRAVO !!! Vous avez trouvé en {6-essais} coups")
        break

    essais -=1
```
